```python
game_team_stats_df = pd.read_csv('./game_teams_stats.csv')
game_team_stats_df.head()
```

```python
game_team_stats_df['pp%'] = game_team_stats_df['powerPlayGoals'] / game_team_stats_df['powerPlayOpportunities']
game_team_stats_df.sort_values(by='pp%', ascending=False)

game_team_stats_df['is_home'] = game_team_stats_df['HoA'] == 'home' # game_team_stats_df[game_team_stats_df['HoA'] == 'home']
game_team_stats_df['is_away'] = game_team_stats_df['HoA'] != 'home' 

game_team_stats_df = game_team_stats_df.sort_values(['team_id', 'game_id'], ascending=[True, True])
game_team_stats_df['is_win'] = np.where(game_team_stats_df['won']==True, 1, 0)

# fix issue with continuous teams
wins = (game_team_stats_df.is_win != game_team_stats_df.shift().is_win).cumsum()
game_team_stats_df["win_streak"] = game_team_stats_df.is_win.groupby(wins).cumsum()

game_team_stats_df
# pd.get_dummies(game_team_stats_df['head_coach'])
```
