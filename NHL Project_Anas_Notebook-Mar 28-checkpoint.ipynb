```python
# Import the data set
goalie = pd.read_csv('game_goalie_stats.csv')
goalie.head()
```

```python
goalie.info()
```

```python
#checking for total null values
goalie.isnull().sum()
```

```python
# Analyze the missing values
goalie.describe(include=[np.object])
```

```python
# Clean the data for the missing decision (win/lose) 
goalie_cleaned = goalie.dropna(how = 'any', subset=['decision'])
```

```python
goalie_cleaned
```

```python
# lets replace Win and Lose by 1 and 0
goalie_cleaned = goalie_cleaned.replace(['L','W'],[0,1])
```

```python
# understanding the rate rate by save percentage
decision = goalie_cleaned.groupby(['decision'])
```

```python
decision['decision'].count()
```

```python
goalie['save_range'] = pd.cut(x=goalie['savePercentage'],bins=[70,72,74,76,78,80,82,84,86,88,90,92,94,96,98,100])
goalie_cleaned.head()
```

```python
goalie_cleaned['save_range']
saves = goalie_cleaned.groupby('save_range',as_index = False).decision.mean()
saves
```

```python
#let's see the correlation between winning and the save percentage
goalie_cleaned['decision'].corr(goalie_cleaned['savePercentage'])
```

```python
ranges=['70','72','74','76','78','80','82','84','86','88','90','92','94','96','98']
saves['ranges']=ranges
saves
```

```python
#visualization:
plt.bar(saves['ranges'], saves['decision'], 
        width= 0.1, color = 'gray', alpha = 0.7, label='decision')
```

```python
shorthanded=goalie_cleaned.groupby('shortHandedSaves',as_index = False).decision.mean()
```

```python
shorthanded
```

```python
#let's see the correlation between winning and the save percentage
goalie_cleaned['decision'].corr(goalie_cleaned['shortHandedSaves'])
```
